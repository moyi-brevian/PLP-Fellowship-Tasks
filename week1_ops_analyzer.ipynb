# Operational Efficiency Tracker
**Industry: Hospital Wards**

This notebook evaluates the performance of three hospital wards by calculating efficiency scores and assigning status flags.

We'll work through four steps:
1. Define a function to calculate efficiency
2. Define a function to classify that efficiency into a status
3. Model our ward data as a list of dictionaries
4. Loop through the data and print a formatted report

## 1. KPI Function: `calculate_efficiency`

Efficiency tells us what percentage of the target output was actually achieved:

`efficiency (%) = (actual_output / target_output) * 100`

We round to two decimal places to keep the output readable.

```python
def calculate_efficiency(actual_output, target_output):
    """
    Returns efficiency as a percentage.

    Parameters:
        actual_output (float): What was actually produced.
        target_output (float): What was planned or expected.

    Returns:
        float: Efficiency percentage rounded to 2 decimal places.
    """
    return round((actual_output / target_output) * 100, 2)
```

```python
def get_operational_status(efficiency):
    """
    Classifies an efficiency score into a status label.

    Parameters:
        efficiency (float): The efficiency percentage to evaluate.

    Returns:
        str: One of 'Critical', 'Warning', or 'Normal'.
    """
    if efficiency < 70:
        return "Critical"
    elif efficiency < 90:
        return "Warning"
    else:
        return "Normal"
```

## 3. Data Model: Hospital Wards

Each ward is a dictionary with three keys:
- `name`: identifies the site
- `actual_output`: patients processed during the period
- `target_output`: planned patient throughput for the same period

All three wards sit in a list so we can loop over them cleanly.

```python
sites = [
    {"name": "Ward A - General Medicine", "actual_output": 112, "target_output": 120},
    {"name": "Ward B - Pediatrics",       "actual_output": 58,  "target_output": 85},
    {"name": "Ward C - Surgical",         "actual_output": 97,  "target_output": 100},
]
```

## 4. Execution: Loop and Report

For each site we:
1. Extract values from the dictionary
2. Calculate efficiency
3. Get the status label
4. Print a formatted summary

`enumerate` gives us a counter to number each report. The separator line keeps things easy to scan.

```python
for i, site in enumerate(sites, start=1):
    name       = site["name"]
    actual     = site["actual_output"]
    target     = site["target_output"]
    efficiency = calculate_efficiency(actual, target)
    status     = get_operational_status(efficiency)

    print(f"{'=' * 40}")
    print(f"Site {i}: {name}")
    print(f"  Actual Output  : {actual} patients")
    print(f"  Target Output  : {target} patients")
    print(f"  Efficiency     : {efficiency}%")
    print(f"  Status         : {status}")

print(f"{'=' * 40}")
```
